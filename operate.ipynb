```python
import os

# 制作flist文件
data_path = 'datasets/test_chart/exp_512'
label_path = 'datasets/test_chart/label_512'

orig_files = os.listdir(data_path)
data_files = list(map(lambda x: os.path.join(data_path, x+'\n'), orig_files))
label_files = list(map(lambda x: os.path.join(label_path, x+'\n'), orig_files))

file_len = len(data_files)
train_len = int(file_len*0.9)
test_len = int(file_len*0.1)
train_data_files = data_files[:train_len]
train_label_files = label_files[:train_len]
test_data_files = data_files[-test_len:]
test_label_files = label_files[-test_len:]

with open('train_data.flist', 'w+') as f:
    f.writelines(train_data_files)
with open('train_label.flist', 'w+') as f:
    f.writelines(train_label_files)
with open('test_data.flist', 'w+') as f:
    f.writelines(test_data_files)    
with open('test_label.flist', 'w+') as f:
    f.writelines(test_label_files)
```
